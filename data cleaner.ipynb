```python
import pandas as pd
import re

df = pd.read_excel("data/Deelnemers 18-11.xlsx")
df.head()

df["skills"] = (
    df["skills"]
    .fillna("")
    .str.lower()
    .str.replace("•", " ")
    .str.replace("–", " ")
    .str.replace("-", " ")
    .str.replace("/", " ")
    .str.replace(",", " ")
    .str.replace("  ", " ")
    .str.replace("|", " ")
    .str.replace("(", "")
    .str.replace(")", "")
)
```

SYNONYMS

```python
synonyms = {
    # -------------------------
    # MANAGEMENT
    # -------------------------
    "management": "management",
    "managment": "management",
    "people management": "people management",
    "leiderschap": "leadership",
    "leadership": "leadership",

    # -------------------------
    # PROJECT / PROGRAM / CHANGE
    # -------------------------
    "projectmanagement": "project management",
    "project management": "project management",
    "project-management": "project management",
    "program management": "program management",
    "programmamanagement": "program management",
    "change management": "change management",
    "changemanagement": "change management",
    "verandermanagement": "change management",

    # -------------------------
    # AGILE / SCRUM / LEAN
    # -------------------------
    "agile": "agile",
    "agile coach": "agile",
    "scrum": "scrum",
    "scrum master": "scrum",
    "scrummaster": "scrum",
    "scrum-master": "scrum",
    "lean": "lean",
    "lean coach": "lean",

    # -------------------------
    # COACHING
    # -------------------------
    "coach": "coaching",
    "coaching": "coaching",
    "coachen": "coaching",
    "mentoring": "coaching",

    # -------------------------
    # CONSULTANCY
    # -------------------------
    "consultant": "consultancy",
    "consultancy": "consultancy",
    "consulting": "consultancy",

    # -------------------------
    # ANALYTISCH
    # -------------------------
    "analytisch": "analytical",
    "analytisch denkvermogen": "analytical",

    # -------------------------
    # INFORMATION / ICT
    # -------------------------
    "ict governance": "ict governance",
    "information management": "information management",
    "document management": "document management",

    # -------------------------
    # COMMUNICATIE
    # -------------------------
    "communicatie": "communication",
    "communication": "communication",
    "stakeholdermanagement": "stakeholder management",
    "stakeholder management": "stakeholder management"
}
```

```python
def normalize_skill(skill):
    s = skill.strip().lower()
    s = re.sub(r"[^a-zA-Z0-9 ]", " ", s)   # rare tekens weg
    s = re.sub(r"\s+", " ", s)            # dubbele spaties weg

    # apply synonyms
    return synonyms.get(s, s)

def clean_skill_string(raw):
    # splitsen op alles wat op separators lijkt
    parts = re.split(r"[;,•\-_/]", raw.lower())
    parts = [normalize_skill(p) for p in parts]
    parts = [p for p in parts if len(p) > 1]
    return list(set(parts))  # uniek maken

def clean_dataframe(df):
    df = df.copy()
    df["skills"] = df["skills"].fillna("").astype(str)

    cleaned = []

    for raw in df["skills"]:
        parts = raw.lower()
        parts = re.split(r"[;,•\-_/]", parts)
        parts = [normalize_skill(p) for p in parts]
        parts = [p for p in parts if len(p) > 1]  # weg met noise
        cleaned.append(list(set(parts)))  # unieke set

    df["clean_skills"] = cleaned
    return df

df = pd.read_excel("data/Deelnemers 18-11.xlsx")
df_clean = clean_dataframe(df)
df_clean.to_excel("data/schoon.xlsx", index=False)
df["skills"] = df["skills"].fillna("").astype(str)
df["clean_skills"] = df["skills"].apply(clean_skill_string)

df[["First Name", "Last Name", "clean_skills"]].head(10)
print("✔ Schoon bestand opgeslagen: data/schoon.xlsx")
```
